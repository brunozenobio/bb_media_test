```python
import json
import pandas as pd

with open("./database/series.json","r",encoding="utf-8") as series:
    datos_serie = json.load(series)

with open("./database/channels.json","r",encoding="utf-8") as series:
    datos_channels = json.load(series)


df_movies = pd.read_csv("./database/movies.csv")
```

```python
df_movies
```

```python
print(f"Hay un total de {len(df_movies)} peliculas y un total de {len(df_movies["genre"].unique())} generos")
```

```python
df_serie = pd.DataFrame(columns=['title',"genre","classification","synopsis","temporadas"
                                 ,"url"])
for serie in datos_serie:
    df_serie.loc[len(df_serie)] = [serie["title"],serie["genre"],serie["classification"],serie["synopsis"],len(serie["seasons"]),serie["url"]]
    

df_serie
```

```python
print(f"Hay un total de {len(df_serie)} series con {len(df_serie["genre"].unique())} generos")


max_temporadas = int(df_serie["temporadas"].idxmax())
min_temporadas = int(df_serie["temporadas"].idxmin())
print(f"La serie con mayor cantidad de temporadas es {df_serie.iloc[max_temporadas]["title"]} con {df_serie.iloc[max_temporadas]["temporadas"]} temporadas" )

```

```python

episodios = pd.DataFrame(columns=['title', 'season', 'number_of_episodes'])

for serie in datos_serie:
    title = serie["title"]  
    for temporada in serie["seasons"]:
        season = temporada["season"]  
        number_of_episodes = len(temporada["episodes"])

        nueva_fila = pd.DataFrame({
            'title': [title],
            'season': [season],
            'number_of_episodes': [number_of_episodes]
        })

        episodios = pd.concat([episodios, nueva_fila], ignore_index=True)


episodios
```

### Top 10 series con mas episodios

```python
episodios.groupby("title")["number_of_episodes"].sum().reset_index().sort_values(by='number_of_episodes',ascending=False).head(10)
```

```python
programas = pd.DataFrame(columns=['name', 'numero_programas'])

for canal in datos_channels:
    name = canal["name"]  
    cantidad_programas = len(canal["programas"])
    nueva_fila = pd.DataFrame({"name":[name],"numero_programas":[cantidad_programas]})
    
    programas = pd.concat([programas, nueva_fila], ignore_index=True)


programas
```
